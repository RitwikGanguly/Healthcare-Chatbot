```python
import requests

def find_nearby_hospitals(latitude, longitude):
    # API endpoint
    url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    # API parameters
    params = {
        "location": f"{latitude},{longitude}",
        "radius": 5000,  # Search radius in meters
        "type": "hospital",
        "key": "AIzaSyDSk0PNdaUakNl5slNMlkSO74H4qJ4KRvI"  # Replace with your actual API key
    }

    # Send a GET request to the API
    response = requests.get(url, params=params)
    data = response.json()
    print(data)
    # Check if the API request was successful
    if response.status_code == 200 and data["status"] == "OK":
        # Process the results
        hospitals = data["results"]
        for i, hospital in enumerate(hospitals[:5]):
            print(f"{i+1}. {hospital['name']}: {hospital['vicinity']}")
    else:
        print("Failed to retrieve nearby hospitals.")

# Example usage
latitude = 22.5754  # Example latitude
longitude = 88.4798  # Example longitude
find_nearby_hospitals(latitude, longitude)
```
